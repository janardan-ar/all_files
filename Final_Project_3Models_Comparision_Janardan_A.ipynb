The dataset is a financial transaction dataset used for fraud detection analysis. It contains transaction-related attributes such as transaction type, transaction amount, sender and receiver balances, and fraud indicators. The primary objective of this dataset is to classify whether a transaction is fraudulent (isFraud = 1) or legitimate (isFraud = 0). Since the target variable is categorical with two classes, this problem is considered a binary classification problem.

Following three classification models used for the dataset:
##### 1  Logistic Regression
A simple and efficient baseline classification model used to predict whether a transaction is fraudulent or legitimate.
##### 2  XGBoost Classifier
An advanced boosting algorithm widely used for fraud detection because of its high accuracy and ability to handle imbalanced datasets effectively.
##### 3  Random Forest Classifier
An ensemble machine learning model that uses multiple decision trees to improve prediction accuracy and handle complex transaction patterns.
