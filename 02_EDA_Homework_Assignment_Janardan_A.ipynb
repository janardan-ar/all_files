### Task 3: Univariate Analysis
- Plot the distribution of the `sepal_length` column.
- Count the unique values in the `species` column and plot a bar chart.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data['sepal_length'])
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
# Your code here
import matplotlib.pyplot as plt

data['species'].value_counts().plot(kind='bar')

plt.title("Count of Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

### Task 4: Bivariate Analysis
- Create a scatter plot of `sepal_length` vs `sepal_width` colored by species.
- Generate a pairplot for the dataset using Seaborn.

```python
# Your code here
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    x='sepal_length',
    y='sepal_width',
    hue='species',
    data=data,
    s=80
)

plt.title("Sepal Length vs Sepal Width (by Species)")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(data, hue='species')

plt.show()
```

### Task 5: Grouping and Aggregation
- Calculate the average `sepal_length` for each species.
- Find the maximum `petal_width` for each species.

```python
# Your code here
grouped = data.groupby('species')
grouped['sepal_length'].mean()
```

```python
grouped['petal_width'].max()
```

### Task 6: Visualizations
- Create a histogram for `petal_length`.
- Generate a box plot for `petal_width` by species.

```python
# Your code here
#import matplotlib.pyplot as plt

plt.hist(data['petal_length'], bins=7)

plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Petal Length")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10, 6))

sns.boxplot(
    x='species',
    y='petal_width',
    data=data,
    hue='species',      # add this
    palette='Set2',
    ax=ax
)

ax.set_title('Distribution of Petal Width by Species', fontsize=14)
ax.set_xlabel('Species', fontsize=12)
ax.set_ylabel('Petal Width (cm)', fontsize=12)

plt.tight_layout()
plt.show()
```
